```python
# problem 2 Four-function Calculator

firstNumber = float(input("Enter first number:"))
secondNumber = float(input("Enter second number:"))

print("1...Addition")
print("2...Subtraction")
print("3...Multiplication")
print("4...Division")

operation = int(input("Enter choice:"))

if operation == 1 :
    print(firstNumber + secondNumber)
    
elif operation == 2 :
    print(firstNumber - secondNumber)
    
elif operation == 3 :
    print (firstNumber * secondNumber)
    
elif operation == 4 :
    
    if secondNumber == 0:
        print("Unable to divid by 0")
    else :
        print (firstNumber / secondNumber)
        
else :
    print ("Invalid menu option")
```
